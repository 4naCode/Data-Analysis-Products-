```python
#karena data memiliki nili nan pada semua baris jadi perlu di hapus karena tidak bisa menyimpulkan 
#atau tidak ada patokan berdasarkan data berbeda dengan indek 251 dan 252 di sekitarnya masih terdapat data
# Indeks yang akan dihapus
indeks = [30, 39, 161, 173]
# Menghapus baris berdasarkan indeks
df = df.drop(index=indeks)
```

```python
plt.figure(figsize=(10,5))
grouped_data = df.groupby('Origin')['Make'].value_counts()
ax = grouped_data.plot(kind='bar', stacked=True)
ax.bar_label(ax.containers[0])
# Menambahkan label pada sumbu dan judul
plt.xlabel('Origin')
plt.ylabel('Count')
plt.title('Count of Makes by Origin')

#menampilkan grouping
print(grouped_data)
# Menampilkan diagram
plt.show()
```

```python
#sebelumnya kita mengubah type data dan menghapus logo dollar
df['Invoice'] = df['Invoice'].replace('[\$,]', '', regex=True).astype(float)
df['MSRP'] = df['MSRP'].replace('[\$,]', '', regex=True).astype(float)
```

```python
#coba kita cari type sedan berdasarkan drivetrain terbanyak dan mengurutkakn berdasarkan invoice tertinggi
sedanFont = sedan[(sedan['DriveTrain'] == 'Front')].sort_values(by='Invoice', ascending=False)
sedanFont
```

```python
sedanFont.hist(figsize=(15,15), bins=30, color='c')
```

```python
rata_rata_kolom = sedanFont[['MSRP', 'Invoice', 'EngineSize', 'Cylinders', 'Horsepower', 'MPG_City', 'MPG_Highway',  'Weight', 'Wheelbase', 'Length']].mean()

# Menampilkan rata-rata
print("Rata-rata Kolom:")
print (rata_rata_kolom)

rata_rata_kolom.plot(kind='line', color='blue', marker='o')
plt.show()
```
